### Compare June and December data across all years 

```python
from scipy import stats
```

```python
# Filter data for desired months
df
```

```python
# Identify the average temperature for June
june = df.loc[df['date'] == 6,:].dropna()
june["tobs"]
june_ave = june['tobs'].mean()
june_ave
```

```python
# Identify the average temperature for December
dec = df.loc[df['date'] == 12,:].dropna()
dec['tobs']
dec_ave = dec['tobs'].mean()
dec_ave
```

```python
# Create collections of temperature data
```

```python
# Run paired t-test, null hypotesis jun_ave = dec_ave
# Need to update to scipy 1.6.2 to be able to run the alternative parameter for a two sided t-test
stats.stats.ttest_rel(june['tobs'].head(1405), dec['tobs'], axis=0, nan_policy='propagate', alternative='two-sided')
```

### Analysis

# This p-value suggests that the probability of a Type I error when you reject the null hypothesis is 2.48e-186 so basically 0. Since this is below the maximum alpha risk of 5% established for this analysis, the null hypothesis should be rejected. The correct conclusion is that the temperature in June and December are not statistically the same. So if you like the heat go in June to Hawaii
